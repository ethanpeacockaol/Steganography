```python
import matplotlib.pyplot as plt
import PIL.Image as img
import numpy as np
```

```python
# read in test image
plot_img = img.open("plot_channels.jpg")
```

```python
# display image dimensions
plot_img
```

```python
# convert to matrix
img_arr = np.array(plot_img)
```

```python
img_arr.shape
```

```python
print("jpg only 3 channels")
```

```python
#extract rgb channels
red = img_arr[:,:,0]
green = img_arr[:,:,1]
blue = img_arr[:,:,2]
```

```python
# reshape rgb  matrixs to 1d  vector
red = red.ravel()
green = green.ravel()
blue = blue.ravel()


# get x axis
x_axis = [x for x in range(len(red))]
```

```python
# plot channels
plt.plot(x_axis, red, label='red')
plt.plot(x_axis, green, label='green')
plt.plot(x_axis, blue , label='blue')

plt.show()
```

```python
# reduce to first 50 pixels
red = red[:50]
green = green[:50]
blue = blue[:50]
x_axis = x_axis[:50]

# plot channels
plt.plot(x_axis, red, 'r', label='red')
plt.plot(x_axis, green, 'g', label='green')
plt.plot(x_axis, blue, 'b', label='blue')

plt.legend()
plt.show()
```
